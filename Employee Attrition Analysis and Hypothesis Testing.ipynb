The percentage of employee leaving the company is 3% having experience of 2 years. Is it same for employees having more than 2 years experience?

```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
sample1 = employee.sample(100, random_state = 0)
sample1
```

```python
count = sample1["left"].value_counts()
count
```

```python
nobs = len(sample1['left'])
nobs
```

```python
value = 0.03
```

```python
stat, p = proportions_ztest(count = count, nobs = nobs, value = value, alternative = "two-sided")
```

```python
p
```

```python
# p is less than 0.05 so reject null hyp
```

Checking whether, the mean avg monthly hours of a employee having experience of 2-5 yrs is the same as that for employee having exp 6 - 10 yrs.

```python
e1 = df[(df.timeSpent_company >= 2) & (df.timeSpent_company <= 5)]
e2 = df[(df.timeSpent_company >= 6) & (df.timeSpent_company <= 10)]
```

```python
print(e1.shape)
print(e2.shape)
```

```python
s1 = e1.sample(25,random_state = 0)
s2 = e2.sample(25,random_state = 0)
```

```python
import scipy
```

```python
scipy.stats.ttest_ind(s1["avgMonthlyHours"],s2["avgMonthlyHours"])
```

```python
#p value > 0.05 so accept
```

```python
# F test
```

```python
from scipy.stats import f
F  = s1.avgMonthlyHours.var()/s2.avgMonthlyHours.var()
F  # test statistics value
```

```python
# degree of freedom

df1 = len(s1) - 1
df2 = len(s2) - 1
```

```python
scipy.stats.f.cdf(F, df1, df2) #p value
```

Hence, there is no significant difference in the variances of the two groups.
